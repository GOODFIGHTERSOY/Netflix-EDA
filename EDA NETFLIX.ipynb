## Date and year and month

```python
df = pd.read_csv("C:/Users/KRITHIK/Documents/EDA PROJECT/Data/netflix_titles.csv")
```

```python
df['date_added'] = df['date_added'].astype(str).str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
```

```python
df.head()
```

## Genre


```python
df['genre']=df['listed_in'].str.split(",").str[0]
```

```python
df.head()
```

## Visualization

## 1)Movies VS Tv Shows

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x='type', order=df['type'].value_counts().index)
plt.title('Count by Type (Movie vs TV Show)')

plt.show()
```

# 2) Top 10 genres

```python
plt.figure(figsize=(10,5))
df['genre'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres (first listed)')
plt.ylabel('Count')
plt.show()
```

## 3) Rating distribution

```python
plt.figure(figsize=(10,4))
df['rating'].value_counts().head(12).plot(kind='bar')
plt.title('Top Ratings')
plt.show()
```

## 4) Release year distribution

```python
sns.histplot(df['release_year'].dropna().astype(int),bins=30)
plt.title('Distribution of Release Years')
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df['description'])

wc = WordCloud(width=800, height=400).generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
